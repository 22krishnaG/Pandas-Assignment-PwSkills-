Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans: Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots that can be created using Seaborn along with their uses:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They are effective in identifying patterns, correlations, and clusters in the data.
Function: seaborn.scatterplot()
Bar Plot:

Use: Bar plots are suitable for comparing categorical data or showing the distribution of a categorical variable. They are helpful in visualizing the count or average of categories.
Function: seaborn.barplot()
Histogram:

Use: Histograms display the distribution of a continuous numerical variable. They help in understanding the frequency or density of data within specific bins or intervals.
Function: seaborn.histplot() or seaborn.displot() for a more flexible interface.
Box Plot:

Use: Box plots provide a visual summary of the distribution of data. They show the median, quartiles, and potential outliers, making it easy to compare the spread of multiple datasets.
Function: seaborn.boxplot()
Heatmap:

Use: Heatmaps are useful for visualizing the correlation between multiple variables in a tabular dataset. They use color intensity to represent the strength of the relationship between variables.
Function: seaborn.heatmap()
These are just a few of the many plots available in Seaborn. The library also offers violin plots, pair plots, joint plots, and many other types of visualizations to suit various data analysis and exploration needs. Seaborn simplifies the process of creating visually appealing and informative plots, making it a popular choice for data visualization tasks in Python.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

Ans: To plot a line plot using the "fmri" dataset from Seaborn, we first need to load the dataset using the seaborn.load_dataset() function. Then, we can use the seaborn.lineplot() function to create the line plot.

Here's how you can do it:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set the title and labels
plt.title("FMRI Dataset - Signal vs Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

In this code, we use sns.load_dataset("fmri") to load the "fmri" dataset from Seaborn. Then, we use sns.lineplot() to create the line plot. We set "timepoint" as the x-axis, "signal" as the y-axis, and use the "event" and "region" columns to differentiate the lines based on different events and regions. The data=fmri_data argument specifies the dataset to be used.

Finally, we add a title and labels to the plot using plt.title(), plt.xlabel(), and plt.ylabel(), respectively. The resulting line plot will show how the "signal" varies with "timepoint" for different events and regions in the "fmri" dataset.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

Ans To plot two box plots using the "titanic" dataset from Seaborn, with 'pclass' on the x-axis and 'age' and 'fare' on the y-axis, you can follow the steps below:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot for 'age' column
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Create the second box plot for 'fare' column
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Adjust layout to prevent overlapping of titles
plt.tight_layout()

# Show the plots
plt.show()
```

In this code, we use sns.load_dataset("titanic") to load the "titanic" dataset from Seaborn. Then, we create two subplots side by side using plt.subplot(1, 2, 1) and plt.subplot(1, 2, 2).

In the first subplot, we use sns.boxplot() to create the box plot for 'age' on the y-axis and 'pclass' on the x-axis. We set the title, xlabel, and ylabel for this subplot.

In the second subplot, we use sns.boxplot() again to create the box plot for 'fare' on the y-axis and 'pclass' on the x-axis. We set the title, xlabel, and ylabel for this subplot as well.

The plt.tight_layout() function is used to adjust the layout and prevent overlapping of titles. Finally, we display both box plots using plt.show(). The resulting plots will show the distribution of age and fare for different passenger classes on the Titanic.

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans: To plot a histogram for the 'price' column of the "diamonds" dataset from Seaborn, using the 'cut' column as the hue, you can follow the steps below:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot the histogram with 'price' column and 'cut' column as hue
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack", kde=True)

# Set the title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

In this code, we use sns.load_dataset("diamonds") to load the "diamonds" dataset from Seaborn. We then use sns.histplot() to create the histogram. We set 'price' as the x-axis, 'cut' as the hue, multiple="stack" to stack the histograms for different cuts, and kde=True to display the kernel density estimate along with the histogram bars.

Finally, we add a title and labels to the plot using plt.title(), plt.xlabel(), and plt.ylabel(). The resulting histogram will show the distribution of diamond prices based on different cuts.

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Ans: To plot a pair plot for the "iris" dataset from Seaborn, using the "species" column as the hue, you can use the sns.pairplot() function. The pairplot() function creates a grid of scatterplots, showing the relationships between different pairs of numerical variables in the dataset.

Here's how you can do it:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create the pair plot with hue as "species"
sns.pairplot(data=iris_data, hue="species")

# Show the plot
plt.show()
```

In this code, we use sns.load_dataset("iris") to load the "iris" dataset from Seaborn. Then, we use sns.pairplot() to create the pair plot. We set hue="species" to use the "species" column as the hue, which will color the points in the scatterplots based on the different species of iris.

The resulting pair plot will show scatterplots for all combinations of numerical variables in the "iris" dataset, and points will be colored based on the species they belong to. This visualization helps to understand the relationships and distributions of different variables in the dataset, especially when considering the species as a categorical variable.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Ans: To plot a heatmap for the "flights" dataset from Seaborn, you can use the sns.heatmap() function. The "flights" dataset contains the number of passengers traveling on various airlines over different months and years.

Here's how you can create the heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_matrix = flights_data.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")

# Set the title and labels
plt.title("Flights Dataset - Number of Passengers")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```

In this code, we use sns.load_dataset("flights") to load the "flights" dataset from Seaborn. Then, we pivot the data using the pivot() method to create a matrix for the heatmap. The resulting matrix will have months as rows, years as columns, and the number of passengers as the values.

Next, we use sns.heatmap() to create the heatmap. We set annot=True to display the values in each cell of the heatmap, fmt="d" to format the values as integers, and cmap="YlGnBu" to use the "YlGnBu" color map.

Finally, we add a title and labels to the plot using plt.title(), plt.xlabel(), and plt.ylabel(). The resulting heatmap will show the number of passengers traveling on different airlines over various months and years. The color intensity will represent the values, making it easy to identify patterns and trends in the data.
